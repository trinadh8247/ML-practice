```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
population = np.random.exponential(scale=2, size=100000)
sample_means = []
checkpoints = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

for i in range(1000):
  sample_means.append(np.mean(np.random.choice(population,30)))
fig, axes = plt.subplots(2, 5, figsize=(20, 8))

for ax, n in zip(axes.flatten(), checkpoints):
    sns.histplot(sample_means[:n], kde=True, color="blue", ax=ax)
    ax.set_title(f"After {n} sample means")

plt.tight_layout()
plt.show()
```
