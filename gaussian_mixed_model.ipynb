```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# Generate synthetic dataset (3 clusters)
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=0.8, random_state=42)

# Fit GMM with 3 components
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm.fit(X)

# Predict cluster labels (hard assignment)
labels = gmm.predict(X)

# Predict cluster probabilities (soft assignment)
probs = gmm.predict_proba(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=40, alpha=0.7)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c="red", marker="X", s=200)
plt.title("Gaussian Mixture Clustering (sklearn)")
plt.show()

# Print parameters
print("Mixing coefficients (weights):", gmm.weights_)
print("Means:\n", gmm.means_)
print("Covariances:\n", gmm.covariances_)
```
